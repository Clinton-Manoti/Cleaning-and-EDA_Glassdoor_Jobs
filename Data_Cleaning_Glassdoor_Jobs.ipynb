# Cleaning the Data

##### Remove all -1s in the DataFrame

```python
# Salary parsing
# Remove all -1's
df_modified = df_modified[df_modified["Salary Estimate"]!='-1']
```

```python
df_modified
```

#### Remove (Glassdoor est.) from the Salary Estimate column

```python
#df_modified["Salary Estimate"] = [x.split("(")[0] for x in df_modified["Salary Estimate"]]
df_modified["Salary Estimate"]= df_modified["Salary Estimate"].apply(lambda x: x.split("(")[0])
```

```python
df_modified
```

#### Remove all 'K' characters and dollar sign in the Salary Estimate column

```python
#df_modified["Salary Estimate"]= [x.replace("K", '').replace("$", '') for x in df_modified["Salary Estimate"]]
df_modified["Salary Estimate"] = df_modified["Salary Estimate"].apply(lambda x: x.replace("K", '').replace("$", ""))
```

```python
df_modified
```

```python
df_modified["Salary Estimate"].dtypes
```

#### Create a new column called 'hourly' 

```python
#df_modified["hourly"] = [1 if 'per hour' in x.lower() else 0 for x in df_modified["Salary Estimate"]]
df_modified["hourly"] = df_modified["Salary Estimate"].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
df_modified.tail(60)
```

#### Create a new column called 'employer provided' from Salary Estimate

```python
#df_modified["employer provided"] = [1 if 'employer provided salary:' in x.lower() else 0 for x in df_modified["Salary Estimate"]]
df_modified["employer provided"] = df["Salary Estimate"].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0) 
```

```python
df_modified.tail(60)
```

```python
df_modified["hourly"]
```

#### Remove all 'per hour' and 'employer provided salary:' words in Salary Estimate

```python
#df_modified["Salary Estimate"] = [x.lower().replace('per hour', '').replace('employer provided salary:', '') for x in df_modified["Salary Estimate"]]
df_modified["Salary Estimate"] = df_modified["Salary Estimate"].apply(lambda x: x.lower().replace('per hour', '').replace("employer provided salary:", ''))
```

```python
df_modified.tail(60)
```

#### Create a new column called 'min salary' to show minimum salary

```python
#df_modified['min_salary'] = [x.split('-')[0] for x in df_modified["Salary Estimate"]]
df_modified["min_salary"] = df_modified["Salary Estimate"].apply(lambda x: x.split("-")[0])
```

```python
df_modified
```

#### Create a 'max salary' column to show maximum salary

```python
df_modified["max_salary"]= [x.split('-')[1] for x in df_modified["Salary Estimate"]]
```

```python
df_modified.head()
```

#### Convert the 'min salary' column into the interger data type 

```python
df_modified["min_salary"] = df_modified["min_salary"].astype('int')
```

#### Convert the 'max salary' column into the interger data type

```python
df_modified["max_salary"] = df_modified["max_salary"].astype('int')
```

#### Find the average salary

```python
df_modified["avg_salary"] = (df_modified.min_salary+df_modified.max_salary)/2
```
